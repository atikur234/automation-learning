```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load a lightweight, high-performance model
# 'all-MiniLM-L6-v2' is the "Swiss Army Knife" of small models.
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
documents = [
    "To make a great beef curry, you need slow-cooked onions and authentic spices.",
    "Python is a versatile programming language used heavily in AI and Data Science.",
    "The traffic in Dhaka can be challenging during peak office hours.",
    "Vector databases allow for high-speed retrieval of high-dimensional data.",
    "Machine learning models require clean data to perform accurately.",
    "A popular snack in Bangladesh is Fuchka, often served with tamarind water."
]
```

```python
document_embeddings = model.encode(documents)
print(f"Shape of embeddings: {document_embeddings.shape}")
```

```python
query = "Who won the World Cup?"
query_embedding = model.encode([query])
similarities = cosine_similarity(query_embedding, document_embeddings)[0]
```

```python
top_indices = np.argsort(similarities)[-2:][::-1]

print(f"Query: {query}\n")
for idx in top_indices:
    print(f"Score: {similarities[idx]:.4f} | Doc: {documents[idx]}")
```
